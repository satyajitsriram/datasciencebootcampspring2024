##### 4)Create a 4x4 NumPy array filled with random integers between 1 and 100. Then, reshape this array into two separate 2D arrays, where one represents the rows and the other represents the columns. Write a function, preferably using a lambda function, to calculate the sum of each row and each column separately, and return the results as two separate NumPy arrays

```python
# Create a 4x4 NumPy array filled with random integers between 1 and 100
numpy4x4 = np.random.randint(1, 101, size=(4, 4))

# Reshape into two separate 2D arrays for rows and columns
rows_array = numpy4x4.reshape(2, 8)
columns_array = numpy4x4.T.reshape(2, 8)

# Function to calculate the sum of each row and column separately using lambda functions
sum_row = lambda array: np.sum(array, axis=1)
sum_column = lambda array: np.sum(array, axis=0)

# Calculate the sums of rows and columns separately
sums_of_rows = sum_row(rows_array)
sums_of_columns = sum_column(columns_array)
```

```python
print("Sums of Rows:", sums_of_rows)
print("Sums of Columns:", sums_of_columns)
```
